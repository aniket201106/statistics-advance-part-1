What is a random variable in probability theory?
  In probability theory, a random variable is a variable that takes on different numerical values based on the outcome of a random experiment.

 What is the difference between discrete and continuous distributions

Discrete Random Variable

Takes on a countable number of values (e.g., 0, 1, 2, 3,...).

Example: Number of heads when flipping 3 coins.

Continuous Random Variable

Takes on an infinite number of possible values within a given range.

Example: Time it takes to complete a task (could be any real number like 2.35 seconds).

What are point estimates and interval estimates in statistics
  In statistics, both point estimates and interval estimates are methods used to infer population parameters (like the population mean or proportion) based on sample data.

What is the significance of confidence intervals in statistical analysis
  A confidence interval (CI) is a range of values that is used to estimate an unknown population parameter with a certain level of confidence (e.g., 90%, 95%, 99%).

 What is the relationship between a Z-score and a confidence interval
  A Z-score is used to construct a confidence interval when the population standard deviation Ïƒ is known and the data follows a normal distribution (or sample size is large enough by the Central Limit Theorem).

```python
#+> Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
from scipy.stats import bernoulli

def bernoulli_pdf(p):
    x = [0, 1]
    pdf = bernoulli.pmf(x, p)
    return list(zip(x, pdf))

print("Bernoulli PDF for p=0.6:", bernoulli_pdf(0.6))
```

```python
#Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
non_normal_data = np.random.exponential(scale=2, size=10000)
sample_means = []

for _ in range(1000):
    sample = np.random.choice(non_normal_data, 30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
#Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
import numpy as np
import matplotlib.pyplot as plt

means = []
for _ in range(1000):
    sample = np.random.normal(loc=50, scale=10, size=30)
    means.append(np.mean(sample))

plt.hist(means, bins=30, edgecolor='k')
plt.title("Central Limit Theorem: Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()

```

```python
#Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
from scipy.stats import norm

def z_score(x, mean, std):
    return (x - mean) / std

x = 72
mean = 70
std = 4

z = z_score(x, mean, std)
print("Z-score:", z)
print("Standard normal CDF:", norm.cdf(z))  # Compare with standard normal
```

```python
#Implement hypothesis testing using Z-statistics for a sample dataset
from scipy.stats import norm

sample_mean = 105
pop_mean = 100
std_dev = 10
n = 30

z = (sample_mean - pop_mean) / (std_dev / np.sqrt(n))
p_value = 1 - norm.cdf(z)

print("Z-statistic:", z)
print("P-value:", p_value)
```

```python
#Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy.stats import norm

data = np.random.normal(100, 15, 50)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

conf = 0.95
z = norm.ppf((1 + conf) / 2)
margin = z * (std / np.sqrt(n))

print(f"Confidence Interval: ({mean - margin:.2f}, {mean + margin:.2f})")
```

```python
#Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
from scipy.stats import norm

data = np.random.normal(100, 15, 50)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

conf = 0.95
z = norm.ppf((1 + conf) / 2)
margin = z * (std / np.sqrt(n))

print(f"Confidence Interval: ({mean - margin:.2f}, {mean + margin:.2f})")
```

```python
#5 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
from scipy.stats import iqr

data = np.random.normal(100, 20, 100)
data_range = np.max(data) - np.min(data)
data_iqr = iqr(data)

print("Range:", data_range)
print("IQR:", data_iqr)
```

```python
# Implement Z-score normalization on a dataset and visualize its transformation
def z_normalize(data):
    return (data - np.mean(data)) / np.std(data)

data = np.random.normal(60, 10, 100)
normalized = z_normalize(data)

print("Mean of normalized data:", np.mean(normalized))
print("Std of normalized data:", np.std(normalized))
```
